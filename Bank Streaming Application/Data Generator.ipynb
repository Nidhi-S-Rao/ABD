```python
import random
from datetime import datetime,date,time
from time import sleep
import json
from google.colab import drive
drive.mount('/content/drive')
```

```python
account_number=[1,2,3,4,5]
transaction=["c","d"]
count=1
for x in range(20):
  readings=[]
  for i in range(10):
    for j in range(5):
      dict={}
      dict["Acc No"]=account_number[random.randint(0,4)]
      dict["Transactions"]=transaction[random.randint(0,1)]
      dict["time"]=str(datetime.now())
      dict["Amount"]=random.randint(1,10000)
      readings.append(dict)
    sleep(1)
  filename="Bank"+str(count)+".json"
  file=open("drive/My Drive/Colab Notebooks/Data/"+filename,"w")
  file.write(json.dumps(readings))
  file.close()
  count+=1
```
