There are few numbers stored in a text file separated with few space and tabs. There is no particular order to store the numbers in a line. A single line can have any number of numbers.

```python
import pyspark
from pyspark import *
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('spark').getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
rdd=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/numbers.txt')
```

```python
rdd.collect()
```

```python
rdd1=rdd.flatMap(lambda d:d.split('\t'))
rdd1.collect()
```

```python
rdd2=rdd1.flatMap(lambda d:d.split(' '))
rdd2.collect()
```

```python
#Method to filter only the numerical values
def method(n):
  if len(n)>0:
    return True
rdd3=rdd2.filter(lambda d:method(d))
```

```python
#Maximum number
rdd3.flatMap(lambda d:[int(d)]).max()
```

```python
#Minimum number
rdd3.flatMap(lambda d:[int(d)]).min()
```

```python
#Sum of the values
rdd3.flatMap(lambda d:[int(d)]).sum()
```

```python
#Mean of the values
rdd3.flatMap(lambda d:[int(d)]).mean()
```
