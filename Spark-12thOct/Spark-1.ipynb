```python
import pyspark
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('spark').getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
#Read a text file
data=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/sampleTemp.txt')
```

```python
data.collect()
```

```python
#Count the number of words in the file using RDD operations.
dataFlatMap=data.flatMap(lambda d:d.split(' '))
dataFlatMap.count()
```

```python
#Write a program to find the word frequency in a given file
dataFlatMap.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
#Write a program to convert all words in a file to uppercase
def isAlphabet(d):
  if d.isalpha():
    return True
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.upper()).collect()
```

```python
#Write a program to convert all words in a file to lowercase.
def isAlphabet(d):
  return d.isalpha()
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.lower()).collect()
```

```python
#Write a program to capitalize first letter of each words in file
def isAlphabet(d):
  return d.isalpha()
dataFlatMap.filter(lambda d:isAlphabet(d)).map(lambda d:d.capitalize()).collect()
```

```python
#Find the number of occurrence of a word in a given file
dataFlatMap.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
#Select only the sentences containing given word from a text file
def patternCheck(d):
  return 'Abohar' in d
data.filter(lambda d:patternCheck(d)).collect()
```

```python
# Find the longest length of word from given set of words
dataFlatMap.map(lambda d:(len(d),d)).max()
```

```python
"""Map the Registration numbers to corresponding branch. 58000 series BDA, 57000 series AIML,
38000 series VLSI, 39000 series ES, and 47000 series CDC. Given registration number, generate a
key-value pair of Registration Number and Corresponding Branch."""

def branch(d):
  if '580' in d:
    return 'BDA'
  elif '570' in d:
    return 'AIML'
  elif '380' in d:
    return 'VLSI'
  elif '390' in d:
    return 'ES'
  elif '470' in d:
    return 'CDC'
students=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/register.txt')
students.flatMap(lambda d:d.split(' ')).map(lambda d:(d,branch(d))).collect()
```

```python
"""Text file contain numbers. Numbers are separated by one white space. There is no order to store
the numbers. One line may contain one or more numbers. Find the maximum, minimum, sum and
mean of numbers"""

numbers=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/numbers.txt')
max=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).max()
min=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).min()
sum=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).sum()
mean=numbers.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).mean()
print(max,min,sum,mean)
```

```python
"""A text file (citizen.txt) contains data about citizens of country. Fields(information in file) are Name,
dob, Phone, email and state name. Another file contains mapping of state names to state code
like Karnataka is codes as KA, TamilNadu as TN, Kerala KL etc. Compress the citizen.txt file by
changing full state name to state code."""


stateRDD=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/codes.txt')
stateKey=stateRDD.map(lambda d:(d.split(' ')[0],d.split(' ')[1]))
stateKey.collect()
codes_dict={}
for val in stateKey.collect():
  codes_dict[val[0]]=val[1]
codes_dict



def changeToCode(d,codes_dict):
  new_data=' '
  split_data=d.split(' ')
  split_data[4]=codes_dict.get(split_data[4])
  new_data=new_data.join(split_data)
  return new_data


citizens=spark.sparkContext.textFile('drive/My Drive/Colab Notebooks/citizens.txt')
mapCitizens=citizens.map(lambda d:changeToCode(d,codes_dict))
mapCitizens.collect()
#mapCitizens.repartition(1).saveAsTextFile('/modified_citizens.txt') #Saving the updated file in RDD
```
